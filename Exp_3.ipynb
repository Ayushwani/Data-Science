```python
import pandas as pd # Import the pandas library and give it the alias 'pd'

df = pd.read_csv('/content/drive/MyDrive/data.csv') # Now you can use pd to refer to pandas
df
```

```python
print(df.columns)
```

```python
df = pd.get_dummies(df, columns=['housing'])

df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['job']=le.fit_transform(df['job'])
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
size_mapping = {'tertiary':0, 'secondary':1, 'unknown':2}
df['education']=df['education'].map(size_mapping)
df
```

```python
frequency_encoding = df['job'].value_counts().to_dict()
df['job'] =df['job'].map(frequency_encoding)
df
```

```python
mean_encoded = df.groupby('job')['education'].mean().to_dict()
df['job'] = df['job'].map(mean_encoded)
df
```

```python
df['marital'] = df['marital'].astype(str)
df
```

```python
from sklearn.feature_extraction import FeatureHasher
import pandas as pd
marital_list = ['married', 'single', 'divorced', 'married', 'widowed']
marital_list = [{'marital_status': status} for status in marital_list]
hasher = FeatureHasher(input_type='dict', n_features=8)
hashed_features = hasher.transform(marital_list)
df_hashed = pd.DataFrame(hashed_features.toarray(), columns=[f'marital_{i}' for i in range(8)])
print(df_hashed)
```
